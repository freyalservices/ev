```python
import pandas as pd
from sklearn.ensemble import IsolationForest
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
```

```python
# Load the CSV file into a DataFrame
df = pd.read_csv('dataset/random_data.csv')
```

```python
# Extract the numeric columns for anomaly detection
numeric_columns = df.select_dtypes(include='number')
```

```python
# Standardize the numeric columns
scaler = StandardScaler()
numeric_columns_standardized = scaler.fit_transform(numeric_columns)
```

```python
# Train the Isolation Forest model
model = IsolationForest(contamination=0.05, random_state=42)
model.fit(numeric_columns_standardized)
```

```python
# Predict anomalies (1 for normal, -1 for anomaly)
df['anomaly'] = model.predict(numeric_columns_standardized)
```

```python
df.head(25)
```

```python
# Print the anomalies
anomalies = df[df['anomaly'] == -1]
print("Anomalies:")
anomalies
```

```python
# Plot anomalies
plt.scatter(df.index, df['anomaly'], c=df['anomaly'], cmap='viridis')
plt.title('Anomaly Detection')
plt.xlabel('Index')
plt.ylabel('Anomaly Score')
plt.show()
```

```python
import joblib 

# Save the model to a file
joblib.dump(model, 'model.joblib')
```

```python
# Load the saved Isolation Forest model
loaded_model = joblib.load('model.joblib')
```

```python
# Take user input for all columns
user_input = {}
for column in numeric_columns.columns:
    value = float(input(f"Enter value for {column}: "))
    user_input[column] = value
```

```python
# Convert user input to a DataFrame and standardize it
user_input_df = pd.DataFrame([user_input])
user_input_standardized = scaler.transform(user_input_df)
```

```python
# Predict anomaly for the user input
anomaly_label = loaded_model.predict(user_input_standardized)[0]
print(f"Anomaly Label for User Input: {anomaly_label}")
```

```python
# Display the anomaly score as well
anomaly_score = loaded_model.decision_function(user_input_standardized)[0]
print(f"Anomaly Score for User Input: {anomaly_score}")
```
